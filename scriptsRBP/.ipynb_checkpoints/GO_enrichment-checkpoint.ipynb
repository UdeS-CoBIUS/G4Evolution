```python
listRBP = ['FAM120A', 'CSTF2T', 'GTF2F1', 'XRN2', 'SLTM', 'NCBP2', 'FASTKD2',
    'RBFOX2', 'EFTUD2', 'DROSHA', 'XRCC6', 'YBX3', 'QKI', 'TIA1', 'LIN28B',
    'HNRNPL', 'GRWD1', 'TBRG4', 'U2AF2', 'LARP7']

dicoGene = {}
dicoId = {}

with open('/home/anais/Documents/Projet/G4RBPevo/data/eCLIP/toDl/Hsap_geneId.csv') as f:
    content = f.read()
    lines = content.split('\n')
    for l in lines:
        w = l.split('\t')
        if len(w) >2:
            dicoId[w[1]] = w[2]

# pprint(dicoId)
for rbp in listRBP:
    df = pd.read_csv('/home/anais/Documents/Projet/G4RBPevo/data/eCLIP/toDl/'+rbp+
        '_HepG2_transcriptBinded.csv', sep=',', names=['index','chr','trStart',
        'trEnd','strand_x','none1','none2','strand_y','none3','transcriptId'], index_col=0)
    df['transcriptId'] = df['transcriptId'].str[:-1]
    df['geneName'] = df['transcriptId'].map(dicoId)

    dicoGene[rbp] = {'HepG2': list(set(df['geneName']))}

    df = pd.read_csv('/home/anais/Documents/Projet/G4RBPevo/data/eCLIP/toDl/'+rbp+
        '_K562_transcriptBinded.csv', sep=',', names=['index','chr','trStart',
        'trEnd','strand_x','none1','none2','strand_y','none3','transcriptId'], index_col=0)
    df['transcriptId'] = df['transcriptId'].str[:-1]
    df['geneName'] = df['transcriptId'].map(dicoId)

    dicoGene[rbp]['K562'] = list(set(df['geneName']))
```

# GO enrichment with goatools

```python
from genes_ncbi_hsap_proteincoding import GENEID2NT as GeneID2nt_hsap #python /home/anais/software/projectsVp3/ncbi_gene_results_to_python.py -o genes_ncbi_hsap_proteincoding.py data/eCLIP/toDl/Hsap_background.tsv
from goatools.base import download_go_basic_obo
from goatools.base import download_ncbi_associations
from goatools.obo_parser import GODag
from goatools.anno.genetogo_reader import Gene2GoReader
from goatools.goea.go_enrichment_ns import GOEnrichmentStudyNS
```

```python
obo_fname = download_go_basic_obo()
fin_gene2go = download_ncbi_associations()
obodag = GODag("go-basic.obo")
```

```python
#run one time to initialize
mapper = {}

for key in GeneID2nt_hsap:
    mapper[GeneID2nt_hsap[key].Symbol] = GeneID2nt_hsap[key].GeneID
    
inv_map = {v: k for k, v in mapper.items()}
```

```python
#run one time to initialize

# Read NCBI's gene2go. Store annotations in a list of namedtuples
objanno = Gene2GoReader(fin_gene2go, taxids=[9606])
# Get namespace2association where:
#    namespace is:
#        BP: biological_process               
#        MF: molecular_function
#        CC: cellular_component
#    assocation is a dict:
#        key: NCBI GeneID
#        value: A set of GO IDs associated with that gene
ns2assoc = objanno.get_ns2assc()
```

```python
#run one time to initialize
goeaobj = GOEnrichmentStudyNS(
        GeneID2nt_hsap.keys(), # List of mouse protein-coding genes
        ns2assoc, # geneid/GO associations
        obodag, # Ontologies
        propagate_counts = False,
        alpha = 0.05, # default significance cut-off
        methods = ['fdr_bh']) # defult multipletest correction method
```

```python
#run one time to initialize
GO_items = []

temp = goeaobj.ns2objgoea['BP'].assoc
for item in temp:
    GO_items += temp[item]
    

temp = goeaobj.ns2objgoea['CC'].assoc
for item in temp:
    GO_items += temp[item]
    

temp = goeaobj.ns2objgoea['MF'].assoc
for item in temp:
    GO_items += temp[item]
```

```python
#pass list of gene symbols
def go_it(test_genes):
    print(f'input genes: {len(test_genes)}')
    
    mapped_genes = []
    for gene in test_genes:
        try:
            mapped_genes.append(mapper[gene])
        except:
            pass
    print(f'mapped genes: {len(mapped_genes)}')
    
    goea_results_all = goeaobj.run_study(mapped_genes)
    goea_results_sig = [r for r in goea_results_all if r.p_fdr_bh < 0.05]
    GO = pd.DataFrame(list(map(lambda x: [x.GO, x.goterm.name, x.goterm.namespace, x.p_uncorrected, x.p_fdr_bh,\
                   x.ratio_in_study[0], x.ratio_in_study[1], GO_items.count(x.GO), list(map(lambda y: inv_map[y], x.study_items)),\
                   ], goea_results_sig)), columns = ['GO', 'term', 'class', 'p', 'p_corr', 'n_genes',\
                                                    'n_study', 'n_go', 'study_genes'])

    GO = GO[GO.n_genes > 1]
    return GO
```

```python
df = go_it(dicoGene['TIA1']['HepG2']) # test with TIA2 HepG2
```

```python
df['per'] = df.n_genes/df.n_go
```
