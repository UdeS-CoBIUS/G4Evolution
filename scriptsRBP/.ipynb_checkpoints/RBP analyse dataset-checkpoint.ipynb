# Get homology grp from the RBP2GO DB

```python
G = nx.Graph()

dicoSpNameCorrespondance = {"Arabidopsis thaliana": "arabidopsis_thaliana",
        "Caenorhabditis elegans": "caenorhabditis_elegans",
        "Danio rerio": "danio_rerio",
        "Drosophila melanogaster": "drosophila_melanogaster",
        "Escherichia coli": "escherichia_coli_str_k_12_substr_mg1655",
        "Homo sapiens": "homo_sapiens",
        "Leishmania donovani": "",
        "Leishmania mexicana": "",
        "Mus musculus": "mus_musculus",
        "Plasmodium falciparum": "",
        "Saccharomyces cerevisiae": "saccharomyces_cerevisiae",
        "Salmonella Typhimurium": "",
        "Trypanosoma brucei": ""}

for index, row in RBPHomologyDf.iterrows():
    if dicoSpNameCorrespondance[row['Organisms']] != 0:
        if row['UniProt_ID'] not in G:
            G.add_node(row['UniProt_ID'], specie=row['Sp'])
        if row['Entry_Name'] not in G:
                G.add_node(row['Cluster_Members'], specie=row['Organisms'])
        G.add_edge(row['UniProt_ID'], row['Cluster_Members'])
```

```python
#trop long à tourner et pas forcement utile
# pos = nx.spring_layout(G, seed=225)  # Seed for reproducible layout
# nx.draw(G, pos)
# plt.show()
```

```python
cc = nx.connected_components(G)

cpt = 0
dicoOrthoFam = {}
for familly in cc:
    famG = nx.Graph(G.subgraph(familly))
    if len(famG) >= 3:
        dicoOrthoFam[str(cpt)] = list(famG.nodes)
        cpt += 1

print(cpt)
for oN in dicoOrthoFam:
    output = open(path+"HomologyRBP/"+oN+".txt", "w")
    output.write('\n'.join(dicoOrthoFam[oN]))
    output.close()
```
