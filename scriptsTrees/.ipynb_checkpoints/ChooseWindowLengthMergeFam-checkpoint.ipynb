# Choix des tailles de window

```python
grpSp = {'Arc': [], 'Bac': [], 'Euk': [],
         'ArcBac': [], 'ArcEuk': [], 'BacEuk': [],
         'ArcBacEuk': []}
grpSpNormalized = {'Arc': [], 'Bac': [], 'Euk': [],
         'ArcBac': [], 'ArcEuk': [], 'BacEuk': [],
         'ArcBacEuk': []}

trees = {}
treesLength = {}

for grp in grpSp:
    files = subprocess.check_output("ls /home/anais/Documents/Projet/G4Evolution/gapFamOrtho/"+grp+"/", shell=True)
    files = files.decode("utf-8")
    for file in files.split('\n'):
        if file != '':
            content = subprocess.check_output('cat /home/anais/Documents/Projet/G4Evolution/gapFamOrtho/'+grp+'/'+file, shell=True)
            content = content.decode("utf-8")
            content = content.split('\n')
            if content[-1] == '':
                content = content[:-1]
            content = [int(x) for x in content]
            grpSp[grp].extend(content)
            trees[file.split('.')[0]] = grp
            treesLength[file.split('.')[0]] = {'Gaps': content}
```

```python
for grp in grpSp:       
    for t in treesLength:
        if grp == trees[t]:
            nbLines = subprocess.check_output("cat /home/anais/Documents/Projet/G4Evolution/ProcessedAlignOrtho/"+grp+"/ProcAlignment_"+t+".csv | wc", shell=True)
            nbLines = nbLines.decode("utf-8")
            try:
                length = int(nbLines.split(' ')[3])
            except:
                try:
                    length = int(nbLines.split(' ')[4])
                except:
                    try:
                        length = int(nbLines.split(' ')[5])
                    except:
                        try:
                            length = int(nbLines.split(' ')[0])
                        except:
                            print(t)
                            print(grp)
                            print(nbLines.split(' '))

            treesLength[t].update({'normGaps' : [x / length * 100 for x in treesLength[t]['Gaps']]})

for grp in grpSp:
    files = subprocess.check_output("ls /home/anais/Documents/Projet/G4Evolution/gapFamOrtho/"+grp+"/", shell=True)
    files = files.decode("utf-8")
    for file in files.split('\n'):
        if file != '':
            t = file.split('.')[0]
            grpSpNormalized[grp].extend(treesLength[t]['normGaps'])
                    
# pprint(trees)
# pprint(grpSp)
```

# Get pG4 per fam

```python
def getNbpG4(dico):
    for grp in dico:
        for t in dico[grp]['pG4 per familly']['Trees']:
            dico[grp]['pG4 per familly']['Trees'][t]['nbpG4'] = {}
            pG4Coord = subprocess.check_output("cat /home/anais/Documents/Projet/G4Evolution/coordpG4/coord_"+t+".csv", shell=True)
            pG4Coord = pG4Coord.decode("utf-8")
            pG4Coord = pG4Coord.split('\n')
            pG4Coord = pG4Coord[1:-1]
            for pG4 in pG4Coord:
                pG4 = pG4.split('\t')
                start = int(pG4[-2])
                end = int(pG4[-1])
                for fam in dico[grp]['pG4 per familly']['Trees'][t]['CoordsFam']:
                    if str(fam[0])+':'+str(fam[1]) not in dico[grp]['pG4 per familly']['Trees'][t]['nbpG4']:
                        dico[grp]['pG4 per familly']['Trees'][t]['nbpG4'][str(fam[0])+':'+str(fam[1])] = 0
                    if (fam[0] >= start and start <= fam[1]) or (fam[0] >= end and end <= fam[1]):
                        dico[grp]['pG4 per familly']['Trees'][t]['nbpG4'][str(fam[0])+':'+str(fam[1])] += 1
    for grp in dico:
        for t in dico[grp]['pG4 per familly']['Trees']:
            for fam in dico[grp]['pG4 per familly']['Trees'][t]['nbpG4']:
                dico[grp]['pG4 per familly']['Values'].append(dico[grp]['pG4 per familly']['Trees'][t]['nbpG4'][str(fam)])
    return dico
```

```python
grpSpStat = getNbpG4(grpSpStat)

for grp in grpSpStat:
    print(grp)
    sns.boxplot(grpSpStat[grp]['pG4 per familly']['Values'])
    plt.show()
    try:
        sns.violinplot(data=grpSpStat[grp]['pG4 per familly']['Values'])
        plt.show()
    except:
        print('plop')
```
