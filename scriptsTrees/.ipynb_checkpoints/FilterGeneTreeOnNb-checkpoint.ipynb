```python
import os
import pandas as pd
import seaborn as sns
from pprint import pprint
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

# Import des données

```python
df = pd.read_csv('/home/anais/Documents/Projet/G4Evolution/geneSpByTree.csv', sep='\t')
df.columns = ['Gene', 'Species', 'Kingdom', 'Tree']
df.head()
```

```python
print(len(list(set(df['Gene']))))
print(len(list(set(df['Species']))))
print(len(list(set(df['Kingdom']))))
print(len(list(set(df['Tree']))))
```

```python
dicoNbGeneSp = {'Arch': {'Nb gene' : [], 'Nb Sp': []},
               'Bact': {'Nb gene' : [], 'Nb Sp': []},
               'Euk': {'Nb gene' : [], 'Nb Sp': []},
               'ArcBact': {'Nb gene' : [], 'Nb Sp': []},
               'ArcEuk': {'Nb gene' : [], 'Nb Sp': []},
               'BactEuk': {'Nb gene' : [], 'Nb Sp': []},
               'ArcBactEuk': {'Nb gene' : [], 'Nb Sp': []}}
dfTree = df.groupby('Tree')
for name, group in dfTree:
    grps = list(set(group['Kingdom']))
    if len(grps) == 3:
        dicoNbGeneSp['ArcBactEuk']['Nb gene'].append(len(list(set(group['Gene']))))
        dicoNbGeneSp['ArcBactEuk']['Nb Sp'].append(len(list(set(group['Species']))))
    elif len(grps) == 2:
        if 'Archaea' in grps and 'Bacteria' in grps:
            dicoNbGeneSp['ArcBact']['Nb gene'].append(len(list(set(group['Gene']))))
            dicoNbGeneSp['ArcBact']['Nb Sp'].append(len(list(set(group['Species']))))
        elif 'Archaea' in grps and 'Eukaryota' in grps:
            dicoNbGeneSp['ArcEuk']['Nb gene'].append(len(list(set(group['Gene']))))
            dicoNbGeneSp['ArcEuk']['Nb Sp'].append(len(list(set(group['Species']))))
        elif 'Bacteria' in grps and 'Eukaryota' in grps:
            dicoNbGeneSp['BactEuk']['Nb gene'].append(len(list(set(group['Gene']))))
            dicoNbGeneSp['BactEuk']['Nb Sp'].append(len(list(set(group['Species']))))
    else:
        if 'Archaea' in grps:
            dicoNbGeneSp['Arch']['Nb gene'].append(len(list(set(group['Gene']))))
            dicoNbGeneSp['Arch']['Nb Sp'].append(len(list(set(group['Species']))))
        elif 'Eukaryota' in grps:
            dicoNbGeneSp['Euk']['Nb gene'].append(len(list(set(group['Gene']))))
            dicoNbGeneSp['Euk']['Nb Sp'].append(len(list(set(group['Species']))))
        elif 'Bacteria' in grps:
            dicoNbGeneSp['Bact']['Nb gene'].append(len(list(set(group['Gene']))))
            dicoNbGeneSp['Bact']['Nb Sp'].append(len(list(set(group['Species']))))
# pprint(dicoNbGeneSp)
```

```python

for grp in dicoNbGeneSp:
    print(grp)
    print('\tNombre de gènes')
    sns.boxplot(dicoNbGeneSp[grp]['Nb gene'])
    plt.show()
    sns.violinplot(data=dicoNbGeneSp[grp]['Nb gene'])
    plt.show()
    print("\tNombre d'espèces")
    sns.boxplot(dicoNbGeneSp[grp]['Nb Sp'])
    plt.show()
    sns.violinplot(data=dicoNbGeneSp[grp]['Nb Sp'])
    plt.show()
```

Mes critères pour choisir le nobmre de gènes/espèces par arbre, est principalement de se limitées au nombres dans le dernier quartiles, ou les outliar.


| Group         | Nombre de gènes | Nombre d'espèces |
|---------------|-----------------|------------------|
| Archée        | 13              | 8                |
| Bactérie      | 20              | 9                |
| Eukaryotes    | 20              | 15               |
| Arch-Bact     | 25              | 15               |
| Arch-Euk      | 35              | 30               |
| Bact-Euk      | 30              | 20               |
| Arch-Bact-Euk | 50              | 35               |

```python
dicoFilteredTree = {}
dfTree = df.groupby('Tree')
for name, group in dfTree:
    name = str(name)
    grps = list(set(group['Kingdom']))
    if len(grps) == 3:
        dicoFilteredTree[name] = {'Nb gene' : len(list(set(group['Gene']))),
                                  'Nb Sp': len(list(set(group['Species']))),
                                  'Grp': 'ArcBacEuk'}
    elif len(grps) == 2:
        if 'Archaea' in grps and 'Bacteria' in grps:
            dicoFilteredTree[name] = {'Nb gene' : len(list(set(group['Gene']))),
                                      'Nb Sp': len(list(set(group['Species']))),
                                      'Grp': 'ArcBac'}
        elif 'Archaea' in grps and 'Eukaryota' in grps:
            dicoFilteredTree[name] = {'Nb gene' : len(list(set(group['Gene']))),
                                      'Nb Sp': len(list(set(group['Species']))),
                                      'Grp': 'ArcEuk'}
        elif 'Bacteria' in grps and 'Eukaryota' in grps:
            dicoFilteredTree[name] = {'Nb gene' : len(list(set(group['Gene']))),
                                      'Nb Sp': len(list(set(group['Species']))),
                                      'Grp': 'BacEuk'}
    else:
        if 'Archaea' in grps:
            dicoFilteredTree[name] = {'Nb gene' : len(list(set(group['Gene']))),
                                      'Nb Sp': len(list(set(group['Species']))),
                                      'Grp': 'Arc'}
        elif 'Eukaryota' in grps:
            dicoFilteredTree[name] = {'Nb gene' : len(list(set(group['Gene']))),
                                      'Nb Sp': len(list(set(group['Species']))),
                                      'Grp': 'Euk'}
        elif 'Bacteria' in grps:
            dicoFilteredTree[name] = {'Nb gene' : len(list(set(group['Gene']))),
                                      'Nb Sp': len(list(set(group['Species']))),
                                      'Grp': 'Bac'}
```

```python
dicoFilter = {'Arc': {'Nb gene' : 13, 'Nb Sp': 8},
               'Bac': {'Nb gene' : 20, 'Nb Sp': 9},
               'Euk': {'Nb gene' : 20, 'Nb Sp': 15},
               'ArcBac': {'Nb gene' : 25, 'Nb Sp': 15},
               'ArcEuk': {'Nb gene' : 35, 'Nb Sp': 30},
               'BacEuk': {'Nb gene' : 30, 'Nb Sp': 20},
               'ArcBacEuk': {'Nb gene' : 50, 'Nb Sp': 35}}
```
